```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
dbutils.widgets.removeAll()
```

```python
dbutils.widgets.text('container','raw')
dbutils.widgets.text('storage-name','adlspruebatecnicamvm')
```

```python
container = dbutils.widgets.get('container')
storage_name = dbutils.widgets.get('storage-name')
```

Departamentos

```python
source_dir = "abfss://" + container + "@" + storage_name + ".dfs.core.windows.net/"

Departamentos_data = source_dir + "Departamentos.csv"

schema = "bronze"
catalogo = "catalog_dev"
```

```python
#Esquema Departamentos
Departamentos_schema= StructType([
StructField('IdDepartamento',IntegerType(),nullable=True ),
StructField('NombreDepartamento',StringType(),nullable=True)
])
```

```python
df_Departamentos = spark.read.schema(Departamentos_schema).csv(Departamentos_data,header=True,sep=",")
```

```python
df_Departamentos.limit(5).display()
```

```python
#transformacion de la data formato delta y almacenamiento
df_Departamentos.write.mode('overwrite').format("delta").insertInto(f"{catalogo}.{schema}.TblDepartamentos")
```

```python
%sql
use catalog `catalog_dev`; select * from `bronze`.`TblDepartamentos` limit 100;
```

Empleados

```python
source_dir = "abfss://" + container + "@" + storage_name + ".dfs.core.windows.net/"

Empleado_data = source_dir + "Empleados.csv"

schema = "bronze"
catalogo = "catalog_dev"
```

```python
#Esquema Empleados
Empleados_schema= StructType([
StructField('IdEmpleado',IntegerType(),nullable=True ),
StructField('NombreEmpleado',StringType(),nullable=True),
StructField('Correo',StringType(),nullable=True),
StructField('Salario',FloatType(),nullable=True),
StructField('IdPuesto',IntegerType(),nullable=True )
])
```

```python
df_Empleados = spark.read.schema(Empleados_schema).csv(Empleado_data,header=True,sep=",")
```

```python
df_Empleados.write.mode('overwrite').format("delta").insertInto(f"{catalogo}.{schema}.TblEmpleados")
```

```python
%sql
use catalog `catalog_dev`; select * from `bronze`.`TblEmpleados` limit 100;
```
