```python
import random
import pandas as pd
```

Se crean los Nombres de los Departamentos de una empresa

```python
NombreDepartamento = [
    "Recursos Humanos", "Ventas", "Finanzas", 
    "IT", "Marketing", "Operaciones", "Logística"
]

Departamentos = [
    {"IdDepartamento": i + 1, "NombreDepartamneto": d}
    for i, d in enumerate(NombreDepartamento)
]
```

Se crean los nombres de los diferentes puestos de trabajo para cada departamento de la empresa.

```python
NombrePuestos = [
    "Analista", "Coordinador", "Gerente", "Director", 
    "Auxiliar", "Ingeniero"
]

PuestosTrabajo = []
for dep in Departamentos:
    num_puestos = random.randint(2, 4)  # puestos por departamento
    for _ in range(num_puestos):
        NombrePuesto = random.choice(NombrePuestos)
        PuestosTrabajo.append({
            "IdPuesto": len(PuestosTrabajo) + 1,
            "NombrePuesto": f"{NombrePuesto} de {dep['NombreDepartamento']}",
            "IdDepartamento": dep["IdDepartamento"]
        })


```

Se crean los empleados correspondientes a los diferentes puestos de trabajo de la empresa

```python
nombres = ["Ana", "Juan", "María", "Luis", "Carlos", "Sofía", "Pedro", "Lucía", "Jorge", "Laura"]
apellidos = ["García", "Pérez", "Rodríguez", "López", "Martínez", "Torres", "Gómez", "Hernández"]

Empleados = []
for i in range(50):  # cantidad de empleados
    nombre = f"{random.choice(nombres)} {random.choice(apellidos)}"
    correo = nombre.replace(" ", ".").lower() + "@empresa.com"

    puesto = random.choice(PuestosTrabajo)
    salario = random.randint(20000, 80000)

    Empleados.append({
        "IdEmpleado": i + 1,
        "NombreEmpleado": nombre,
        "Correo": correo,
        "Salario": salario,
        "IdPuesto": puesto["IdPuesto"]
    })
```

Se crean los Dataframe para Departamentos, Puestos de Trabajo y Empleados y se visualizan para su validacion.

```python
df_departamentos = pd.DataFrame(Departamentos)
df_puestos = pd.DataFrame(PuestosTrabajo)
df_empleados = pd.DataFrame(Empleados)

display(df_departamentos)
display(df_puestos)
display(df_empleados)
```

Se lleva cada uno de los Dataframe a un archivo CSV

```python
df_departamentos.to_csv("Departamentos.csv", index=False)
df_puestos.to_csv("PuestosTrabajo.csv", index=False)
df_empleados.to_csv("Empleados.csv", index=False)
```
