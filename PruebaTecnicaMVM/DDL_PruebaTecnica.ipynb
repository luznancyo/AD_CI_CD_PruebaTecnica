```python
%sql
create widget text storageLocation default "abfss://raw@adlspruebatecnicamvm.dfs.core.windows.net/";
create widget text esquema default "bronze";
```

```python
%sql
create catalog if not exists catalog_dev;
```

```python
%sql
create schema if not exists catalog_dev.raw;
create schema if not exists catalog_dev.bronze;
create schema if not exists catalog_dev.silver;
create schema if not exists catalog_dev.golden;
```

```python
%sql
Drop table if exists catalog_dev.${esquema}.TblDepartamentos;
```

```python
%sql
Create or replace table catalog_dev.${esquema}.TblDepartamentos (
    IdDepartamento integer,
NombreDepartamento string
    
)
USING DELTA
LOCATION '${storageLocation}/bronze/TblDepartamentos';
```

```python
%sql
Drop table if exists catalog_dev.${esquema}.TblPuestosTrabajo;
```

```python
%sql

Create or replace table catalog_dev.${esquema}.TblPuestosTrabajo (
 IdPuesto integer,
NombrePuesto string,
IdDepartamento integer
    
)
USING DELTA
LOCATION '${storageLocation}/bronze/TblPuestosTrabajo';
```

```python
%sql
Drop table if exists catalog_dev.${esquema}.TblEmpleados;
```

```python
%sql
Create or replace table catalog_dev.${esquema}.TblEmpleados (
IdEmpleado integer,
NombreEmpleado string,
Correo string,
Salario float,
IdPuesto integer
    
)
USING DELTA
LOCATION '${storageLocation}/bronze/TblEmpleados';
```
